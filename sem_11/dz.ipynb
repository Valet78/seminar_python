# Для функции f(x) = sin(x)^2 - cos(x)^2
---
1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
# import math as mf
import sympy as smp
from sympy.abc import x 

exp_f = smp.expand(smp.sin(x)**2 - smp.cos(x)**2)
exp_f
```

### Определить корни

```python
rts = smp.solve(exp_f, x)
rts
```

### Вычислить вершину

```python
diff_fun = smp.diff(exp_f, x) # дифферинциал функции
diff_fun
```

```python
top = smp.solve(diff_fun, x)
top
```

### Построить график

```python
graf = smp.plot(exp_f, line_color='red', title='График заданной функции')
```

### Найти интервалы, на которых функция возрастает

```python
grow_f = smp.solve(diff_fun > 0)
grow_f
```

### Найти интервалы, на которых функция убывает

```python
drop_f = smp.solve(diff_fun < 0)
drop_f
```

### Определить промежутки, на котором f > 0

```python
pozit = smp.solve(exp_f > 0)
pozit
```

### Определить промежутки, на котором f < 0

```python
negativ = smp.solve(exp_f < 0)
negativ
```
